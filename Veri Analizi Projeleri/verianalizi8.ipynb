```python
import random

categories = ["Elektronik", "Giyim", "Kitap", "Kozmetik", "EvYasam"]

with open("big_orders.txt", "w", encoding="utf-8") as f:
    for i in range(1000):
        order_id = 1000 + i
        
        customer = random.choice(["Ali", "Ayse", "Can", "Elif", "", "Mehmet"])
        
        category = random.choice(categories + ["giyim", "ELEKTRONIK"])
        
        amount = random.choice([120, 450, 900, "", "abc", 1500, 2500])
        
        status = random.choice(["OK", "ERROR"])
        
        row = f"{order_id},{customer},{category},{amount},{status}"
        
        f.write(row + "\n")

print("1000 kayıtlık ham veri oluşturuldu.")
```

```python
missing_customer = 0
bad_amount = 0
bad_category = 0
valid_categories = ["Elektronik", "Giyim", "Kitap", "Kozmetik", "EvYasam"]
with open("big_orders.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        customer = parts[1]
        category = parts[2]
        amount = parts[3]
        if customer == "":
            missing_customer += 1
        if amount == "" or not amount.isdigit():
            bad_amount += 1
        if category not in valid_categories:
            bad_category += 1

print("Eksik müşteri:", missing_customer)
print("Hatalı tutar:", bad_amount)
print("Standart dışı kategori:", bad_category)
```

```python
clean_rows = []
with open("big_orders.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        order_id = parts[0]
        customer = parts[1] if parts[1] != "" else "Bilinmiyor"
        category = parts[2].capitalize()
        amount = parts[3]
        status = parts[4]
        
        if amount == "" or not amount.isdigit():
            amount = "0"
            
        row = f"{order_id},{customer},{category},{amount},{status}"

        clean_rows.append(row)

with open("clean_big_orders.txt", "w", encoding="utf-8") as f:
    for row in clean_rows:
        f.write(row + "\n")

print("Temiz veri hazır.")
```

```python
new_rows = []
with open("clean_big_orders.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        amount = int(parts[3])
        status = parts[4]
        if amount >= 1500:
            level = "Premium"
        elif amount >= 500:
            level = "Standart"
        else:
            level = "Basic"
        is_error = 1 if status == "ERROR" else 0
        
        row = line.strip() + f",{level},{is_error}"
        
        new_rows.append(row)

with open("featured_big_orders.txt", "w", encoding="utf-8") as f:
    for row in new_rows:
        f.write(row + "\n")

print("Feature engineering tamamlandı.")
```

```python
import time

start = time.time()
total = 0
with open("featured_big_orders.txt", "r", encoding="utf-8") as f:
    for line in f:
        total += 1
end = time.time()
print("Tek dosya okuma süresi:", round(end - start, 4), "sn")

start = time.time()
files = {
"Elektronik": 0,
"Giyim": 0,
"Kitap": 0,
"Kozmetik": 0,
"Evyasam": 0
}
with open("featured_big_orders.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        category = parts[2]
        if category in files:
            files[category] += 1
end = time.time()
print("Partition bazlı işlem süresi:", round(end - start, 4), "sn")

print(files)
```

```python
category_counts = {}
premium_count = 0
error_count = 0

with open("featured_big_orders.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        category = parts[2]
        level = parts[5]
        is_error = int(parts[6])
        
        category_counts[category] = category_counts.get(category, 0) + 1
        
        if level == "Premium":
            premium_count += 1
            
        error_count += is_error

print("Kategori dağılımı:")
for k, v in category_counts.items():
    print(k, "->", v)
    
print("Premium sipariş sayısı:", premium_count)

print("Toplam hata:", error_count)
```
