```python
import random
from datetime import datetime

cities = ["Istanbul", "Ankara", "Izmir", "Bursa"]

with open("web_logs.csv", "w", encoding="utf-8") as f:
 f.write("timestamp,city,response_time,status\n")

 for i in range(10000):
     timestamp = datetime.now().isoformat(timespec="seconds")
     city = random.choice(cities)
     response_time = random.randint(50, 500)
     status = "ERROR" if random.random() < 0.05 else "OK"

     f.write(f"{timestamp},{city},{response_time},{status}\n")

print("web_logs.csv oluşturuldu.")
```

```python
total = 0
errors = 0

with open("web_logs.csv", "r", encoding="utf-8") as f:
 next(f)
 for line in f:
     total += 1
     if line.endswith("ERROR\n"):
         errors += 1

print("Toplam kayıt:", total)
print("Toplam hata:", errors)
print("Hata oranı:", round(errors/total*100,2), "%")
```

```python
error_by_city = {}

with open("web_logs.csv", "r", encoding="utf-8") as f:
 next(f)
 for line in f:
     parts = line.strip().split(",")
     city = parts[1]
     status = parts[3]
     if status == "ERROR":
         error_by_city[city] = error_by_city.get(city, 0) + 1
print("Şehirlere göre ERROR sayısı:")
for city, count in error_by_city.items():
     print(city, count)
```

```python
times = []

with open("web_logs.csv", "r", encoding="utf-8") as f:
 next(f)
 for line in f:
     parts = line.strip().split(",")
     times.append(int(parts[2]))
times.sort()

print("Minimum:", times[0])
print("Maximum:", times[-1])
print("Ortalama:", sum(times)/len(times))

p95 = times[int(len(times)*0.95)]

print("P95:", p95)
```

```python
bad_rows = 0
good_rows = 0

with open("web_logs.csv", "r", encoding="utf-8") as f:
    next(f)
    for line in f:
        parts = line.strip().split(",")
        
        if len(parts) != 4:
            bad_rows += 1
            continue 
            
        if not parts[2].isdigit():
            bad_rows += 1
            continue 
            
        good_rows += 1

print("Geçerli kayıt:", good_rows)
print("Hatalı kayıt:", bad_rows)
```

```python
def read_data(file):
    with open(file, "r", encoding="utf-8") as f:
        next(f) 
        for line in f:
            yield line.strip().split(",")

def count_errors(data):
    errors = 0
    total = 0
    for row in data:
        total += 1
        if row[3] == "ERROR":
            errors += 1
    return total, errors

data = read_data("web_logs.csv")
total, errors = count_errors(data)

print("Toplam:", total)
print("ERROR:", errors)
```
