```python
import random
import time
from datetime import datetime

cities = ["Istanbul", "Ankara", "Izmir", "Bursa"]

with open("stream_data.txt", "a", encoding="utf-8") as f:
 for i in range(50):
     timestamp = datetime.now().isoformat(timespec="seconds")
     city = random.choice(cities)
     value = random.randint(1, 100)
     f.write(f"{timestamp},{city},{value}\n") 
     f.flush()
time.sleep(0.2)
```

```python
import time

counts = {}

with open("stream_data.txt", "r", encoding="utf-8") as f:
 f.seek(0, 2)
 while True:
     line = f.readline()
     if not line:
         time.sleep(0.1)
         continue
 
     parts = line.strip().split(",")
     city = parts[1]
     counts[city] = counts.get(city, 0) + 1

     print("Anlık sayım:", counts)
```
