```python
import random

contents = ["Film", "Dizi", "Belgesel", "Muzik", "Podcast"]
devices = ["Mobile", "Web", "TV", "Tablet"]

with open("stream_logs.txt", "w", encoding="utf-8") as f:

    for i in range(500):

        user_id = random.randint(1000, 9999)

        content = random.choice(contents + ["film"])

        watch_time = random.choice([5, 12, 30, 60, "", "abc", 90])

        device = random.choice(devices)

        status = random.choice(["OK", "ERROR"])

        row = f"{user_id},{content},{watch_time},{device},{status}"

        f.write(row + "\n")

print("500 kayıt üretildi.")
```

```python
missing_time = 0
bad_time = 0
bad_content = 0

valid_contents = ["Film", "Dizi", "Belgesel", "Muzik", "Podcast"]

with open("stream_logs.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        watch_time = parts[2]
        if watch_time == "":
            missing_time += 1
        if watch_time != "" and not watch_time.isdigit():
            bad_time += 1
        if content not in valid_contents:
            bad_content += 1

print("Eksik süre:", missing_time)
print("Hatalı süre:", bad_time)
print("Standart dışı içerik:", bad_content)
```

```python
clean_rows = []
with open("stream_logs.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        user_id = parts[0]

        content = parts[1].capitalize()
        watch_time = parts[2]
        device = parts[3]
        status = parts[4]
        
        if watch_time == "" or not watch_time.isdigit():
            watch_time = "0"
            
        row = f"{user_id},{content},{watch_time},{device},{status}"

        clean_rows.append(row)

with open("clean_logs.txt", "w", encoding="utf-8") as f:
    for row in clean_rows:
        f.write(row + "\n")

print("Temiz veri hazır.")
```

```python
new_rows = []
with open("clean_logs.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        watch_time = int(parts[2])
        status = parts[4]
        
        if watch_time >= 60:
            level = "High"
        elif watch_time >= 20:
            level = "Medium"
        else:
            level = "Low"
            
        is_error = 1 if status == "ERROR" else 0
        row = line.strip() + f",{level},{is_error}"
        new_rows.append(row)

with open("featured_logs.txt", "w", encoding="utf-8") as f:
    for row in new_rows:
        f.write(row + "\n")

print("Feature engineering tamamlandı.")
```

```python
total = 0

with open("featured_logs.txt", "r", encoding="utf-8") as f:
    for line in f:
        total += 1

print("Toplam kayıt:", total)


with open("featured_logs.txt", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if i < 5:
            print("Anlık log:", line.strip())
```

```python
device_counts = {}

with open("featured_logs.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        device = parts[3]
        device_counts[device] = device_counts.get(device, 0) + 1

print("Cihaz bazlı dağılım:")

for k, v in device_counts.items():
    print(k, "->", v)
```
