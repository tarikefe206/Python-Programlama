```python
raw_data = [
"1001,Ali,Elektronik,1200,OK",
"1002,Ayşe,Giyim,450,OK",
"1003,,Kitap,150,OK",
"1004,Mehmet,Elektronik,,ERROR",
"1005,Zeynep,Giyim,780,OK",
"1006,Ali,elektronik,950,OK",
"1007,Can,Kitap,abc,OK",
"1008,Elif,Kozmetik,560,OK",
"1009,Ahmet,GIYIM,430,ERROR",
"1010,Fatma,EvYasam,890,OK"
]

with open("raw_orders.txt", "w", encoding="utf-8") as f:
 for row in raw_data:
     f.write(row + "\n")
print("Ham veri dosyası oluşturuldu.")
```

```python
with open("raw_orders.txt", "r", encoding="utf-8") as f:
 for line in f:
     parts = line.strip().split(",")
     order_id = parts[0]
     customer = parts[1]
     category = parts[2]
     amount = parts[3]
     status = parts[4]
     
     print("Kayıt:", parts)

     if customer == "":
         print(" -> UYARI: Müşteri adı eksik")
     if amount == "":
         print(" -> UYARI: Tutar eksik")
     if not amount.isdigit() and amount != "":  
         print(" -> UYARI: Tutar sayısal değil")
     if category not in ["Elektronik", "Giyim", "Kitap", "Kozmetik", "EvYasam"]:
         print(" -> UYARI: Kategori standardı bozuk")

     print("-" * 40)
```

```python
cleaned_data = []
with open("raw_orders.txt", "r", encoding="utf-8") as f:

    for line in f:
     parts = line.strip().split(",")
     order_id = parts[0]
     customer = parts[1] if parts[1] != "" else "Bilinmiyor"
     category = parts[2].capitalize()
     amount = parts[3]
     status = parts[4]

     if category == "Elektronik":
         category = "Elektronik"
     elif category == "Giyim" or category == "Giyim".upper():
         category = "Giyim"
     if amount == "" or not amount.isdigit():
         amount = "0"

     cleaned_row = f"{order_id},{customer},{category},{amount},{status}"
     cleaned_data.append(cleaned_row)

with open("clean_orders.txt", "w", encoding="utf-8") as f:
  for row in cleaned_data:
     f.write(row + "\n")

print("Temiz veri dosyası oluşturuldu.")
```

```python
featured_data = []


with open("clean_orders.txt", "r", encoding="utf-8") as f:
 for line in f:
     parts = line.strip().split(",")
     order_id = parts[0]
     customer = parts[1]
     category = parts[2]
     amount = int(parts[3])
     status = parts[4]

     if amount >= 1000:
         order_size = "Buyuk"

     elif amount >= 500:
         order_size = "Orta"

     else:
         order_size = "Kucuk"


     is_error = 1 if status == "ERROR" else 0


     new_row = f"{order_id},{customer},{category},{amount},{status},{order_size},{is_error}"

     featured_data.append(new_row)


with open("featured_orders.txt", "w", encoding="utf-8") as f:
 for row in featured_data:
     f.write(row + "\n")
     print("Feature engineering tamamlandı.")
```

```python
category_counts = {}
order_size_counts = {}
error_count = 0

with open("featured_orders.txt", "r", encoding="utf-8") as f:

 for line in f:
 
     parts = line.strip().split(",")
     category = parts[2]
     order_size = parts[5]
     is_error = int(parts[6])
     category_counts[category] = category_counts.get(category, 0) + 1
     order_size_counts[order_size] = order_size_counts.get(order_size, 0) + 1
     error_count += is_error
print("Kategori bazlı kayıt sayısı:")

for k, v in category_counts.items():
    print(k, "->", v)
    print("\nSipariş büyüklüğü dağılımı:")

    for k, v in order_size_counts.items():
     print(k, "->", v)
     print("\nToplam hata sayısı:", error_count)
```
