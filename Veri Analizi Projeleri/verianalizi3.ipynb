```python
import random
from datetime import datetime

N = 20000
cities = ["Istanbul", "Ankara", "Izmir", "Bursa"]
statuses = ["OK"] * 19 + ["ERROR"]

with open("server_log.csv", "w", encoding="utf-8") as f:
    f.write("timestamp,city,response_time,status\n")
    for _ in range(N):
        ts = datetime.now().isoformat(timespec="seconds")
        city = random.choice(cities)
        rt = random.randint(50, 600)
        status = random.choice(statuses)
        f.write(f"{ts},{city},{rt},{status}\n")

print("server_log.csv üretildi. Satır sayısı:", N)
```

```python
total = 0
errors = 0

with open("server_log.csv", "r", encoding="utf-8") as f:
    header = next(f)  
    for line in f:   
        total += 1    
        if line.rstrip().endswith(",ERROR"): 
            errors += 1 

rate = round((errors / total) * 100, 2) if total else 0

print("Toplam kayıt:", total)
print("ERROR:", errors)
print("Hata oranı (%):", rate)
```

```python
chunk_size = 5000
buffer = []
chunk_count = 0
error_in_chunks = 0

with open("server_log.csv", "r", encoding="utf-8") as f:
    next(f)
    for line in f:
        buffer.append(line)
        
        if len(buffer) == chunk_size:
            chunk_count += 1
            for row in buffer:
                if row.rstrip().endswith(",ERROR"):
                    error_in_chunks += 1
            buffer.clear()

    if buffer:
        chunk_count += 1
        for row in buffer:
            if row.rstrip().endswith(",ERROR"):
                error_in_chunks += 1

print("Toplam chunk:", chunk_count)
print("Chunk mantığıyla bulunan ERROR:", error_in_chunks)
```

```python
times = []
with open("server_log.csv", "r", encoding="utf-8") as f:
    next(f)
    for line in f:
        parts = line.strip().split(",")
        rt = int(parts[2])
        times.append(rt)

times.sort()
n = len(times)
avg = sum(times) / n if n else 0
p95_index = int(0.95 * (n - 1)) if n else 0

print("Kayıt:", n)
print("Min:", times[0])
print("Max:", times[-1])
print("Ortalama:", round(avg, 2))
print("P95:", times[p95_index])
```

```python
bad_lines = 0
good_lines = 0

with open("server_log.csv", "r", encoding="utf-8") as f:
    next(f)
    for line in f:
        parts = line.strip().split(",")
        
        if len(parts) != 4:
            bad_lines += 1
            continue
            
        ts, city, rt, status = parts
        
        if not rt.isdigit():
            bad_lines += 1
            continue
            
        if status not in ("OK", "ERROR"):
            bad_lines += 1
            continue
            
        good_lines += 1

print("İyi satır:", good_lines)
print("Bozuk satır:", bad_lines)
```

```python
import random
from datetime import datetime

def generate(filename, n=5000):
    cities = ["Istanbul", "Ankara", "Izmir", "Bursa"]
    statuses = ["OK"] * 19 + ["ERROR"]
    with open(filename, "w", encoding="utf-8") as f:
        f.write("timestamp,city,response_time,status\n")
        for _ in range(n):
            ts = datetime.now().isoformat(timespec="seconds")
            f.write(f"{ts},{random.choice(cities)},{random.randint(50,600)},{random.choice(statuses)}\n")

def analyze(filename):
    total = 0
    errors = 0
    with open(filename, "r", encoding="utf-8") as f:
        next(f)
        for line in f:
            total += 1
            if line.rstrip().endswith(",ERROR"):
                errors += 1
    return total, errors

file = "pipeline_log.csv"
generate(file, n=10000)
total, errors = analyze(file)

print("Dosya:", file)
print("Toplam:", total, "ERROR:", errors)
```
