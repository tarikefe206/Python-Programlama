```python
import oyun_modulu

def main():
    print("Modüler Şans Oyunları Uygulamasına Hoş Geldiniz!")
    oyuncu_adi = input("Lütfen oyuncu adınızı giriniz: ").strip()
    
    if not oyuncu_adi:
        oyuncu_adi = "Misafir"

    while True:
        print("\n" + "="*30)
        print("ŞANS OYUNLARI ÖRNEK MENÜ:")
        print("1. Sayı Tahmin")
        print("2. Yazı-Tura")
        print("3. Skorlarım")
        print("4. Çıkış")
        print("="*30)

        try:
            secim = int(input("Lütfen yapmak istediğiniz işlemi seçiniz (1-4): "))
        except ValueError:
            print("Hata: Lütfen geçerli bir menü numarası giriniz!")
            continue

        if secim == 1:
            puan = oyun_modulu.sayi_tahmin()
            oyun_modulu.skor_kaydet(oyuncu_adi, "Sayı Tahmin", puan)
        
        elif secim == 2:
            puan = oyun_modulu.yazi_tura()
            oyun_modulu.skor_kaydet(oyuncu_adi, "Yazı-Tura", puan)
        
        elif secim == 3:
            oyun_modulu.skor_goster()
        
        elif secim == 4:
            print("Çıkış yapılıyor. Oynadığınız için teşekkürler!")
            break
        
        else:
            print("Hatalı seçim! Lütfen 1 ile 4 arasında bir numara giriniz.")

if __name__ == "__main__":
    main()
```

```python
import random

def sayi_tahmin():
    hedef_sayi = random.randint(1, 100)
    hak = 7
    print("\n--- Sayı Tahmin Oyununa Hoş Geldiniz ---")
    print("1 ile 100 arasında bir sayı tuttum. Toplam 7 hakkınız var.")

    while hak > 0:
        try:
            tahmin = int(input(f"Kalan hakkınız {hak}. Tahmininiz: "))
        except ValueError:
            print("Hata: Lütfen sadece sayısal bir değer giriniz!")
            continue 

        if tahmin == hedef_sayi:
            print(f"Tebrikler! {hedef_sayi} sayısını doğru tahmin ettiniz. 50 puan kazandınız!")
            return 50
        elif tahmin < hedef_sayi:
            print("İpucu: Daha büyük bir sayı giriniz.")
        else:
            print("İpucu: Daha küçük bir sayı giriniz.")
        
        hak -= 1

    print(f"Maalesef bilemediniz. Tuttuğum sayı: {hedef_sayi}. 0 puan.")
    return 0

def yazi_tura():
    secenekler = ["yazı", "tura"]
    sonuc = random.choice(secenekler)
    print("\n--- Yazı-Tura Oyununa Hoş Geldiniz ---")
    
    tahmin = input("Tahmininiz nedir? (yazı/tura): ").strip().lower()
    
    if tahmin not in secenekler:
        print("Hatalı giriş yaptınız. Lütfen sadece 'yazı' veya 'tura' yazınız. 0 puan.")
        return 0

    print(f"Para atıldı... Gelen sonuç: {sonuc}")
    
    if tahmin == sonuc:
        print("Tebrikler! Doğru tahmin ettiniz. 20 puan kazandınız!")
        return 20
    else:
        print("Yanlış tahmin. 0 puan.")
        return 0

def skor_kaydet(oyuncu, oyun, puan):
    with open("skorlar.csv", "a", encoding="utf-8") as dosya:
        dosya.write(f"{oyuncu},{oyun},{puan}\n")
    print(f"[{oyun}] skorunuz ({puan} puan) başarıyla kaydedildi.")

def skor_goster():
    print("\n--- Skor Tablosu ---")
    try:
        with open("skorlar.csv", "r", encoding="utf-8") as dosya:
            satirlar = dosya.readlines()
            if not satirlar:
                print("Henüz hiç skor kaydedilmemiş.")
            else:
                print(f"{'Oyuncu':<15} | {'Oyun':<15} | {'Puan':<5}")
                print("-" * 40)
                for satir in satirlar:
                    bilgiler = satir.strip().split(',')
                    if len(bilgiler) == 3:
                        print(f"{bilgiler[0]:<15} | {bilgiler[1]:<15} | {bilgiler[2]:<5}")
    except FileNotFoundError:
        print("Hata: 'skorlar.csv' dosyası bulunamadı. Yeni bir skor dosyası oluşturuluyor...")
        with open("skorlar.csv", "w", encoding="utf-8") as dosya:
            pass 
        print("Dosya oluşturuldu. Henüz gösterilecek bir skor yok.")
```
