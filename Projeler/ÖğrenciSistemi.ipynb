```python
class Ogrenci:
    def __init__(self,isim,numara,bolum):
        self.isim = isim
        self.numara = numara
        self.bolum = bolum
        self.notlar = []
    
    def not_ekle(self, not_degeri):
        self.notlar.append(not_degeri)

    def ortalama_hesapla(self):
        if self.notlar:
            return sum(self.notlar) / len(self.notlar)
        return 0 
    
    def bilgileri_goster(self):
        print(f"İsim: {self.isim}")
        print(f"Bölüm: {self.bolum}")
        print(f"Notlar: {self.notlar}")
        print(f"Ortalama: {self.ortalama_hesapla()}\n")
```

```python
ogr1 = Ogrenci("Tarık", "30", "Veri Analistliği")
ogr2 = Ogrenci("Sena", "51", "Bilgisayar Mühendisliği")

ogr1.not_ekle(80)
ogr2.not_ekle(90)
```

```python
ogr1.bilgileri_goster()
ogr2.bilgileri_goster()
```
