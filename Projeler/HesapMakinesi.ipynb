```python
def Topla(x,y):
    return x+y

def Cikar(x,y):
    return x-y

def Carp(x,y):
    return x*y 

def Bol(x,y):
    return x/y    
```

```python
def hesapla(sayi1, oprt, sayi2):
    if oprt == "+":
        return Topla(sayi1, sayi2)
    elif oprt == "-":
        return Cikar(sayi1, sayi2)
    elif oprt == "*":
        return Carp(sayi1, sayi2)
    elif oprt == "/":
        return Bol(sayi1, sayi2)
    else:
        return "Yanlış Karakter Girdiniz"
```

```python
sayi1 = float(input("1. Sayı: "))
oprt = input("İşlem: ")
sayi2 = float(input("2. Sayi"))

sonuc = hesapla(sayi1, oprt, sayi2)
print(sonuc)
```

```python
print("E Harfi İle Programdan Çıkabilirsiniz")
sonuc = float(input("1. Sayi"))
while True:
    oprt = input("İşlem: ")
    if oprt == "e":
        break
    sayi2 = float(input("2. Sayi"))

    sonuc = hesapla(sayi1, oprt, sayi2)
    print(sonuc)
```
