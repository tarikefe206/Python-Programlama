```python
# Çalışan Mail
it_mailleri = []

with open ('data/calisanlar.txt','r', encoding='utf-8') as dosya:
    for satir in dosya:
        satir = satir.strip()
        parcalar = satir.split(', ')

        isim_soyisim = parcalar[0]
        departman = parcalar[1]

        if departman == 'IT':

            isim, soyisim = isim_soyisim.split()
            email = f"{isim.lower()}.{soyisim.lower()}@sirket.com"
            it_mailleri.append(email)
            print(email)
```

```python
with open('it_mailleri.txt', 'w', encoding='utf-8') as dosya:
    for mail in it_mailleri:
        dosya.write(mail + '\n')
        print(f"\n{len(it_mailleri)} IT Çalışanının E-Postası Kaydedildi.")
```

```python
log_sayilari = {'INFO': 0, 'WARNING': 0, 'ERROR': 0, 'CRITICAL': 0}

with open('data/server.log', 'r', encoding='utf-8') as giris, \
     open('errors.log', 'w', encoding='utf-8') as error_dosya, \
     open('critical.log', 'w', encoding='utf-8') as critical_dosya:

    for satir in giris: # Satır satır oku
        # Log seviyesini belirle
        for seviye in log_sayilari.keys():
            if seviye in satir:
                log_sayilari[seviye] += 1
                
                # Kritik logları filtrele
                if seviye == 'ERROR':
                    error_dosya.write(satir)
                elif seviye == 'CRITICAL':
                    critical_dosya.write(satir)
                
                break

# Rapor oluştur
print("LOG ANALİZ RAPORU")
print("="*40)

for seviye, sayi in log_sayilari.items():
    print(f"{seviye}: {sayi:,} adet")

print(f"\nERROR ve CRITICAL loglar ayrı dosyalara kaydedildi.")
print(f"Toplam {sum(log_sayilari.values()):,} log satırı analiz edildi.")
```
